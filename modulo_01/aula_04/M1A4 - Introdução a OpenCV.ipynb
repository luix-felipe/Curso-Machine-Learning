```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Ler Imagens do Disco

Na aula anterior vimos como ler imagens do disco utilizando matplotlib, mas também podemos fazer utilizando OpenCV.

OpenCV tem uma vantagem por possuir funções para lidar com dispositivos e vídeos.

Para isso utilizamos a função `cv2.imread`

```python
# TODO Atualize o path da imagem.
image = cv2.imread("praticas/modulo_1/aula_4/bahia.jpeg")
```

Aqui é possível observar a importância de `numpy` já que `OpenCV` utiliza a a mesma estrutura de dados.

```python
print(type(image))
```

```python
plt.imshow(image)
```

## Transformações de Imagens

A utilização da mesma estrutura de dados da `numpy` permite realizar as mesmas operações.

```python
gain = np.clip(image /255 * 1.001, 0, 1)
plt.imshow(gain)
```

As imagens em `OpenCV` são lidas em `BGR` ao invés de `RGB` para visualizar da maneira correta com `matplotlib` precisamos converter o espaço de cores.

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gain_rgb = np.clip(image_rgb /255 * 1.001, 0, 1)
plt.imshow(gain_rgb)
```

```python
crop = image_rgb[0:500, 0:500]
plt.imshow(crop)
```

Porém `OpenCV` também implementa algumas operações úteis

```python
larger = cv2.resize(image_rgb, (int(image.shape[1] * 4), int(image.shape[0] * 4)))
plt.imshow(larger)
```
