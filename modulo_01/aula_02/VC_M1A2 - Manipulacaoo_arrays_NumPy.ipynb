## Operações Básicas com Vetores

Primeiro vamos criar vetores para que possamos entender como eles são representados, e como podemos manipular eles. 

Nesse exemplos, vamos utilizar os vetores $\vec{u}$ e $\vec{v}$ com coordenadas $(1, 3)$ e $(3, 1)$ respectivamente como ilustrado na imagem abaixo.

![vetores u e v](vetores.png)

```python
# Criando vetores.
u = np.array([1, 3])
v = np.array([3, 1])
```

Operações feitas individualmente nas dimensões do vetores como substração e adição podem ser feitas utilizando os operadores $-$ e $+$.

```python
y = u - v
w = u + v

print(f"Vetor y: {y}\nVetor w: {w}")
```

Analogamente podemos visualizar esses resultados no plano cartesiano.

![subtração e adição com u e v](sub-add-vecs.png)

Multiplicação e divisão são feitas elemento por elemento.

Com essa multiplicação podemos realizar o produto escalar, por exemplo $\alpha * \vec{u}$

```python
a = u / v
b = u * v

print(f"Vetor a: {a}\nVetor b: {b}")

alpha = 2
escalar = alpha * u
print(f"Produtor escalar de u por alpha: {escalar}")
```

Também é possível realizar operações geométrica como produto interno e vetorial.

```python
interno = np.dot(u, v)


# Para o produto vetorial iremos utilizar vetores tridimensionais.
u3 = np.array([1, 3, 1])
v3 = np.array([3, 1, 1])
vetorial = np.cross(u3, v3)

print(f"Resultado produto interno: {interno}\nResutlado produto vetorial: {vetorial}")
```

## Operações Básicas com Matrizes

A primeira coisa a compreender é como representar matrizes. Analogamente a vetores, em NumPy matrizes são vetores bidimesionais.

Vamos fazer alguns exemplos e observar que podemos reproduzir todas as operações.

```python
I = np.array([[1, 0],[0, 1]])
A = np.array([[2, 1],[1, 2]])
```

De maneira semelhante quando utilizamos vetores as operações podem ser realizadas elemento a elemento, ou seguindo a algebra que estamos familiarizados.

```python
B = A + I

print(f"Matriz A:\n{A}\n\nMatriz Identidade:\n{I}\n\nMatriz B = A + I:\n{B}")
```

```python
B = A - I

print(f"Matriz A:\n{A}\n\nMatriz Identidade:\n{I}\n\nMatriz B = A - I:\n{B}")
```

Os operadores não precisam nem ter exatamente as mesmas dimensões.

```python
C = A + u

print(f"Matriz A:\n{A}\nVetor u:\n{u}\n\nMatriz C = A + u:\n{C}")
```

Para realizarmos a multiplicação de matrizes também podemos utilizar NumPy. Desde que observe-se a regra de colunas e linhas compatíveis entre os operadores.

```python
D = np.matmul(A, I)

print(f"Matriz A:\n{A}\n\nMatriz Identidade:\n{I}\n\nMatriz D = A x I:\n{D}")
```
